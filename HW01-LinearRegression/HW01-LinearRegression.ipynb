## Problem 1 (30 points)

Given observations $\bX\;$ and  $\by$, and using equation (10) in lecture 3, we see that the fitted values $\bhy\,$ are given by
$$
\bhy = \bX \cdot ( \bX\;^T \cdot \bX\;)^{-1} \cdot \bX\;^T  \cdot \by
$$
We define the $n\times n$ matrix operator
$$
P := \bX \cdot ( \bX\;^T \cdot \bX\;)^{-1} \cdot \bX\;^T \qquad (1)
$$
and its complement
$$
M := \bid - P \qquad (2)
$$
with $\bid\,$ the $n \times n$ identity matrix.  
The matrix $P$ is an **orthogonal projection** matrix. It projects any vector $\by\,$ into $P\cdot \by$, which lives in the subspace spanned by the column vectors of $\bX$.  

Show the above statement, by proving the following:  
1) $P^2=P$ (projecting twice is the same as projecting once, idempotent matrix)  
2) $(P\cdot\bx)^T \cdot (M \cdot \by)= 0$ (the projections of $P$ and $M$ are orthogonal vectors for every $\bx$ and $\by$)

## Solution 1

1) Proof:
It is easy to prove that
$$
\begin{align}
P^2& = \bX \cdot ( \bX\;^T \cdot \bX\;)^{-1} \cdot \bX\;^T\cdot \bX \cdot ( \bX\;^T \cdot \bX\;)^{-1} \cdot \bX\;^T\\
& =  \bX \cdot ( \bX\;^T \cdot \bX\;)^{-1} \cdot (\bX\;^T\cdot \bX\;) \cdot ( \bX\;^T \cdot \bX\;)^{-1} \cdot \bX\;^T\\
 &= \bX \cdot( \bX\;^T \cdot \bX\;)^{-1} \cdot \bX\;^T\\
& = P\;.
\end{align}
$$

2) Proof:
First, we can easily prove that the orthogonal projection matrix $P$ is also symmetric as follows:
$$
\begin{align}
P^T & = \left(\bX \cdot ( \bX\;^T \cdot \bX\;)^{-1} \cdot \bX\;^T\right)^T\\
& = \bX \cdot ( \bX\;^T \cdot \bX\;)^{-1} \cdot \bX\;^T\\
& = P\;.
\end{align}
$$
Then we can obtain that 
$$
\begin{align}
(P\cdot\bx\;)^T \cdot (M \cdot \by\;) & = \bx^T\cdot P^T\cdot (I-P)\cdot \by\\
& = \bx^T\cdot (P^T- P^T\cdot P)\cdot \by\\
& = \bx^T\cdot (P- P^2)\cdot \by & (\mbox{since P is symmtric matrix})\\
& = 0\;.& ({\text{since $P = P^2$}})\\
\end{align}
$$

## Solution 2
### import data

```python
# import libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# import the dataset  
dataset = pd.read_csv('stock-treasury-2004_2006.csv')
dataset.head()
```

### calculate excess return for stocks and SP index

```python
# function to calculate tock and SP index daily returns
def asset_daily_return(item):
    return [(item[i+1] / item[i] -1)*10000 for i in range(len(item)-1)]

# function to calculate treasury daily return
def treasury_daily_return():
    TREAS_3M = dataset.loc[:, 'TREAS_3M'].values
    return [(TREAS_3M[i] / 90) * 100 for i in range(len(TREAS_3M))] # 3M treasury note in percent 

# calculate excess returns
treasury_dr = treasury_daily_return() # treasury daily return
items_er = []
for i in range(2, 13):
    item = dataset.iloc[:, i].values
    item_dr = asset_daily_return(item)
    item_er = [item_dr[i] - treasury_dr[i] for i in range(len(item_dr))] # calculate excess return for each item 
    items_er.append(item_er)

# create new dataframe for excess returns
items_er = np.transpose(np.array(items_er))
df_items_er = pd.DataFrame(items_er, columns = ['GM_er', 'F_er', 'UTX_er', 'CAT_er', 'MRK_er', 'PFE_er', 'IBM_er', 'MSFT_er', 'C_er', 'XOM_er', 'SP_er'])

# summarize the new dataframe for excess returns 
df_items_er.describe() 
```

### linear regression using sk-learn

```python
from sklearn.linear_model import LinearRegression

SP_er = df_items_er.loc[:, 'SP_er'].values.reshape(-1, 1)
syms =  ['GM', 'F', 'UTX', 'CAT', 'MRK', 'PFE', 'IBM', 'MSFT', 'C', 'XOM']
items_index = []
for i in range(0, 10):
    stock_er =  df_items_er.iloc[:, i].values
    regressor = LinearRegression()
    regressor.fit(SP_er, stock_er)
    alpha = np.round(regressor.intercept_, 4) 
    beta = float(np.round(regressor.coef_, 4)) 
    R_square = np.round(regressor.score(SP_er, stock_er), 4) 
    sym = syms[i]
    items_index.append([sym, alpha, beta, R_square])
items_index = np.array(items_index)
df_items_index = pd.DataFrame(items_index, columns = ['sym', 'alpha', 'beta', 'r_squared'])

df_items_index
```

### linear regression using statsmodels

```python
import statsmodels.api as sm
# This handles the intercept. 
# Statsmodel takes 0 intercept by default
SP_er = df_items_er.loc[:, 'SP_er'].values
SP_er = sm.add_constant(SP_er)
syms =  ['GM', 'F', 'UTX', 'CAT', 'MRK', 'PFE', 'IBM', 'MSFT', 'C', 'XOM']
items_index = []
for i in range(0, 10):
    stock_er =  df_items_er.iloc[:, i].values
    model = sm.OLS(stock_er, SP_er)
    results = model.fit() 
    alpha = np.round(results.params[0], 4) 
    alpha_se = np.round(results.bse[0], 4) 
    alpha_pval = np.round(results.pvalues[0], 4) 
    beta = np.round(results.params[1], 4)
    beta_se = np.round(results.bse[1], 4) 
    beta_pval = np.round(results.pvalues[1], 4) 
    R_squared = np.round(results.rsquared, 4)
    sym = syms[i]
    items_index.append([sym, alpha, alpha_se, alpha_pval, beta, beta_se, beta_pval, R_squared])
items_index = np.array(items_index)
df_items_index_stat = pd.DataFrame(items_index, columns = ['sym', 'alpha', 'alpha_se', 'alpha_pval', 'beta', 'beta_se', 'beta_pval', 'r_squared'])
    
df_items_index_stat
```
